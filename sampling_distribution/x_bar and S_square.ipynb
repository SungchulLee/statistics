```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_samples = 5
n_simulations = 10000
samples = np.random.normal(size=(n_simulations, n_samples))
x_bar = samples.mean(axis=1)
plt.hist(x_bar, bins=30, density=True)
plt.plot([0,0], [0,1], '--r')
plt.show()
```

$$
S^2=\frac{\sum_{i=1}^n(X_i - \bar{X})^2}{n-1}
$$

```python
n_samples = 10
n_simulations = 10000
samples = np.random.normal(size=(n_simulations, n_samples))
x_bar = samples.mean(axis=1).reshape((-1,1)) # (10,) ---> (10,1)
S_square = np.sum((samples - x_bar)**2, axis=1) / (n_samples-1) 

plt.hist(S_square, bins=30, density=True)
plt.plot([1,1], [0,1], '--r')
plt.show()
```
