$$\begin{array}{ccc}
   &\mbox{goodness of fit}&\mbox{independence}&\\
H_0&\mbox{dist}(X)=\mbox{benchmark dist}&X,\ Y\ \mbox{independent}\\
H_a&\mbox{dist}(X)\not=\mbox{benchmark dist}&X,\ Y\ \mbox{not independent}\\
\mbox{Data}&x\sim X&(x,y)\sim (X,Y)\\
\mbox{scipy}&\mbox{scipy.stats.chisquare}&\mbox{scipy.stats.chi2_contingency}
\end{array}$$

```python
import numpy as np
import scipy.stats as stats

f_obs = np.array([18, 10, 17, 15, 18, 18]) * 4
f_exp = np.array([16, 16, 16, 16, 16, 16]) * 4

chi2, p_value = stats.chisquare(f_obs, f_exp=f_exp)
print(chi2)
print(p_value)
```

```python
import pandas as pd

data = {'obs': f_obs, 'exp': f_exp}
df = pd.DataFrame(data).transpose()
df.columns = [1,2,3,4,5,6]
df.loc['contribution_to_chi2',:] = (f_obs-f_exp)**2/f_exp
print(df)
```

$$
\mbox{DF} = \#\ \mbox{categories} - 1
$$

```python
DF = f_obs.shape[0] - 1
print(DF)
```
