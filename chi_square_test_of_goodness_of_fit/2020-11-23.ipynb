```python
import numpy as np
import scipy.stats as stats
```

```python
O = np.array([110,110,110,90,90,90]) # O.shape = (6,)
E = np.ones_like(O) * 100
```

```python
statistic = ((O-E)**2/E).sum()
print(statistic)

n = O.shape[0] # n = 6
a = stats.chi2(n-1)
p_value = 1 - a.cdf(statistic)
print(p_value)
```

```python
statistic, p_value = stats.chisquare(O, E)
print(statistic)
print(p_value)
```
