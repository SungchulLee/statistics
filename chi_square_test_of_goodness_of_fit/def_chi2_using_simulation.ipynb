$$
\chi^2(d)\sim\sum_{i=1}^dZ_i\quad\mbox{where $Z_i$ are iid $N(0,1)$}
$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
d = 5
n_samples = 1000
z = np.random.normal(size=(d, n_samples))  # (5,3)
chi_square = np.sum(z**2, axis=0)
plt.hist(chi_square, bins=20)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
df = 5
n_samples = 1000
chi_square = np.random.chisquare(df, size=(n_samples,))  # (5,3)
plt.hist(chi_square, bins=20)
plt.show()
```

```python
%matplotlib inline
import scipy.stats as stats
import matplotlib.pyplot as plt
df = 5
n_samples = 1000
a = stats.chi2(df)
chi_square = a.rvs(size=(n_samples,))
plt.hist(chi_square, bins=20, density=True)

x = np.linspace(0, 25, 200)
y = a.pdf(x)
plt.plot(x, y)

plt.show()
```
