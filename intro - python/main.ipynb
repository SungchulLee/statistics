```python
class OnePlusOne:
    
    def __init__(self, string):
        self.string = string
        
    def __add__(self, other):
        if (self.string=='1') and (other.string=='1'):
            return '일더하기 일은 귀요미'
        
a = OnePlusOne('1')
b = OnePlusOne('1')
c = a + b
print(c)
```
