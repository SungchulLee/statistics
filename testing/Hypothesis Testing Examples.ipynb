```python
import numpy as np
import scipy.stats as stats
x = np.array([170,167,174,179,179,187,179,183,179,
              156,163,156,187,156,167,156,174,170,
              183,179,174,179,170,159,187])
n = x.shape[0] 
x_bar = x.mean()
s = x.std(ddof=1)
mu = 170.
t = (x_bar - mu) / (s/np.sqrt(n))
a = stats.t(n-1)
p_value = 1 - a.cdf(t)
print(n)
print(x_bar) # 172.52
print(s) # 10.312290401910397
print(s/np.sqrt(n))
print(t)
print(p_value)
```

```python
import numpy as np
import scipy.stats as stats
x = np.array([11.5,11.8,15.7,16.1,14.1,10.5,9.3,15.0,11.1,
              15.2,19.0,12.8,12.4,19.2,13.5,12.2,13.3,
              16.5,13.5,14.4,16.7,10.9,13.0,10.3,15.8,
              15.1,17.1,13.3,12.4,8.5,14.3,12.9,13.5])
n = x.shape[0] 
x_bar = x.mean()
s = x.std(ddof=1)
mu = 15.7
n_statistic = (x_bar - mu) / (s/np.sqrt(n))
a = stats.norm()
p_value = a.cdf(n_statistic)
print(n)
print(x_bar) # 13.663636363636364
print(s) # 10.312290401910397
print(s/np.sqrt(n))
print(n_statistic)
print(p_value)
```

```python
import numpy as np
import scipy.stats as stats
n_x, n_y = 34, 29
x_bar, y_bar = 105.5, 90.9
s_x, s_y = 20.1, 12.2
n_statistic = (x_bar - y_bar) / np.sqrt(s_x**2/n_x + s_y**2/n_y)
a = stats.norm()
p_value = 2*a.cdf(-np.abs(n_statistic))
print(n_statistic)
print(p_value)
```

```python
import numpy as np
import scipy.stats as stats
n_x, n_y = 34, 29
x_bar, y_bar = 105.5, 90.9
s_x, s_y = 20.1, 12.2
s_p = np.sqrt( ((n_x-1)*s_x**2 + (n_y-1)*s_y**2) / (n_x+n_y-2) )
t_statistic = (x_bar - y_bar) / ( s_p * np.sqrt(1/n_x + 1/n_y) )
a = stats.t( min(n_x-1,n_y-1) )
p_value = 2*a.cdf(-np.abs(t_statistic))
print(t_statistic)
print(p_value)
```
