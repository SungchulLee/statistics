# paired data

```python
import numpy as np
import scipy.stats as stats

d_bar = -14.4
s_d = 6.77
n = 10

# mu_d = 0 <----> mu_d < 0 
t = d_bar / (s_d/np.sqrt(n))
a = stats.t(n-1)
p_value = a.cdf(t)
print(t)
print(p_value)
```

# f test

```python
import numpy as np
import scipy.stats as stats

n_x = 20
n_y = 30
s_x = 100
s_y = 150

# sigma_x = sigma_y <----> sigma_x < sigma_y
f = s_x**2 / s_y**2
a = stats.f(n_x-1,n_y-1)
p_value = a.cdf(f)
print(f)
print(p_value)
```
