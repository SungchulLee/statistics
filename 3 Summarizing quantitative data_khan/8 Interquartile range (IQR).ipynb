```python
import pandas as pd

data = {'x': [4,4,6,7,10,11,12,14,15]}
df = pd.DataFrame(data)
print(df.describe())
```

```python
print(df.quantile(q=0.25))
print(df.quantile(q=0.25, interpolation='nearest'))
```

```python
import scipy.stats as stats

print(stats.iqr(df))
```
