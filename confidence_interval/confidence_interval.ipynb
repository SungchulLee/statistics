$$\begin{array}{lll}
\mbox{Symetric Case}&&\\
\mbox{Confidence_Interval}_\theta 
&=&\hat{\theta}\pm\mbox{Margin_of_Error}_\theta\\
\\
\\
\mbox{Non-Symetric Case}&&\\
\mbox{Confidence_Interval}_\theta 
&=&(\mbox{Left}_\theta, \mbox{Right}_\theta)\\
&=&\mbox{Center}_\theta\pm\frac{\mbox{Right}_\theta-\mbox{Left}_\theta}{2}\\
&=&\mbox{Center}_\theta\pm\mbox{Margin_of_Error}_\theta
\end{array}$$

# Condition for $p$ Interval

$$
\hat{p}\pm z_*\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
$$

$$\begin{array}{lllll}
(1)\ \mbox{Random Sample}\\
(2)\ \mbox{Normal Condition}&&n\hat{p}\ge 10,\quad n(1-\hat{p})\ge 10\\
(3)\ \mbox{Independence Condition}&&n < \mbox{10% of total population if chosen without repetition}\\
\end{array}$$

[Binomial proportion confidence interval](https://en.wikipedia.org/wiki/Binomial_proportion_confidence_interval)

# Condition for $z$ Interval

$$
\bar{X}\pm z_*\frac{S}{\sqrt{n}}
$$

$$\begin{array}{lllll}
(1)\ \mbox{Random Sample}\\
(2)\ \mbox{Normal Condition}&&n\ge 30\quad\mbox{or}\quad\mbox{Original is Normal}\\
\end{array}$$

# Condition for $t$ Interval

$$
\bar{X}\pm t_*(df=n-1)\frac{S}{\sqrt{n}}
$$

$$\begin{array}{lllll}
(1)\ \mbox{Random Sample}\\
(2)\ \mbox{Normal Condition}&&\mbox{Original is Normal}\\
\end{array}$$

[Confidence interval](https://en.wikipedia.org/wiki/Confidence_interval)

```python
import scipy.stats as stats
import numpy as np

n = 500
p_bar = 0.7
s = np.sqrt(p_bar * (1 - p_bar))

z_star = stats.norm.ppf(0.95)

center = p_bar
margin = z_star * s / np.sqrt(n) 
ci_90 = [center - margin, center + margin]
print(ci_90)
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3)
y_normal = stats.norm.pdf(x)
plt.plot(x, y_normal, '--b', label='Normal')

for df in range(1,20):
    y_t = stats.t.pdf(x, df) 
    if df % 5 == 1:
        plt.plot(x, y_t, '--', label=f'T df={df}')
    else:
        plt.plot(x, y_t, '--')
    
plt.legend()
plt.show()
```
